```python
import subprocess
import os
import sys

# ==========================================
# 設定エリア
# ==========================================
input_file = "Shibaura Institute of Technology Toyosu Campus 15.m4a"    # 元のファイル
output_file = "recording_smartphone1.wav" # 保存するファイル
# ==========================================

def convert_to_wav():
    # 1. ファイル確認
    if not os.path.exists(input_file):
        print(f"エラー: '{input_file}' が見つかりません。")
        return

    # 2. コマンド作成
    # -i: 入力, -y: 上書き許可, -vn: 映像なし(音声のみ), -acodec pcm_s16le: 汎用的なWAV形式指定
    command = [
        "ffmpeg",
        "-i", input_file,
        "-y",
        "-vn",
        "-acodec", "pcm_s16le", 
        output_file
    ]

    print(f"--- 変換を開始します ---\n対象: {input_file} -> {output_file}")
    
    try:
        # 3. 実行（ここを変更しました）
        # capture_output=True を削除し、出力を直接ターミナルに流します。
        # これによりフリーズを防ぎ、FFmpegの進捗バーが表示されるようになります。
        result = subprocess.run(command, check=True)

        print("\n✅ 変換完了！")

    except subprocess.CalledProcessError as e:
        print("\n❌ 変換エラーが発生しました。")
        # エラー時はFFmpeg自体が画面に理由を表示してくれます
    except FileNotFoundError:
        print("\n❌ エラー: ffmpeg コマンドが見つかりません。")
        print("WSLで 'sudo apt install ffmpeg' (または conda install ffmpeg) を実行してください。")

if __name__ == "__main__":
    convert_to_wav()
```
