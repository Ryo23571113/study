```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# --- 設定 ---
# 自分のファイルを使う場合はここを書き換えてください
# 例: audio_path = 'data/my_recording.wav'
# ここではlibrosaに付属しているサンプル音源('choice')を使用します
audio_path = '8_deskofwood.wav'

# --- 1. 音声データの読み込み ---
# y: 音声データ（振幅の配列）
# sr: サンプリングレート（1秒間のデータ数）
y, sr = librosa.load(audio_path)

print(f"サンプリングレート: {sr} Hz")
print(f"データ長: {len(y)} サンプル")
print(f"再生時間: {len(y) / sr:.2f} 秒")

# --- 2. グラフの描画 ---
plt.figure(figsize=(12, 6))  # グラフのサイズ設定

# 波形の描画 (librosa.display.waveshowを使うと便利です)
# alphaは透明度
librosa.display.waveshow(y, sr=sr, alpha=0.5)

# --- 3. 軸ラベルとタイトルの設定 ---
plt.title("Audio Waveform")      # タイトル
plt.xlabel("Time (seconds)")     # 横軸：時間
plt.ylabel("Amplitude")          # 縦軸：振幅（音の強さ）

# グリッド線を表示
plt.grid(True)

# グラフを表示
plt.tight_layout()
plt.show()
```
